```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

Image.MAX_IMAGE_PIXELS = None
```

```python
file_path1 = "../../Deforestation-selected/Hansen_GFC-2022-v1.10_last_10N_080W.tif"

img_raw1 = Image.open(file_path1)
img_np1 = np.array(img_raw1)
img_np1.shape
```

```python
plt.imshow(img_np1)

plt.show()
```

```python
file_path2 = "../../Deforestation-selected/Hansen_GFC-2022-v1.10_lossyear_10N_080W.tif"

img_raw2 = Image.open(file_path2)
img_np2 = np.array(img_raw2)
img_np2.shape
```

```python
plt.imshow(img_np2)

plt.show()
```

```python
for i, lat in enumerate(np.linspace(10, 0.5, 20)):
    for j, long in enumerate(np.linspace(-80, -70.5, 20)):
        print(lat, long)
        curr1 = img_np1[(i*2000):((i+1)*2000), (j*2000):((j+1)*2000)]
        cv2.imwrite("../datasets/maps/last/{}_{}.jpg".format(lat, long), curr1)
        curr2 = img_np2[(i*2000):((i+1)*2000), (j*2000):((j+1)*2000)]
        cv2.imwrite("../datasets/maps/lossyear/{}_{}.jpg".format(lat, long), curr2)
```
